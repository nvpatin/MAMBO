# Create Metadata Files

Need to create new metadata files by marker (for merged data) and for the overall dataset

```python
import pandas as pd
# import glob
```

### Set Locations

```python
directory = "./Data/merged_data/"
markers = ['18S', 'COI']
prefixes = ['RREAS2018', 'Flyer2018']
```

# 18S Dataset

## Import Data

```python
marker = '18S'
prefix = 'Merged2018'

# otu table
file = prefix +'_'+marker+'_otu_filtered.csv'
print(directory+file)
df = pd.read_csv(directory + file)
df = df.rename(columns= {'Unnamed: 0': 'ASV'})
df.set_index('ASV', inplace=True)
otu_all = df.copy()
print('Number ASVs:', len(df.index))
```

```python
# Overall Metadata file for Flyer2018 dataset
file = './Data/Flyer2018_metadata.tsv'
df= pd.read_csv(file, sep='\t')
df.set_index('sample_name', inplace=True)
Flyer_meta = df.copy()
df.head()
```

```python
# RREAS metadata
file = './Data/filtered_data/RREAS2018_18S_meta_filtered.csv'
df= pd.read_csv(file, sep=',')
df.set_index('sample_name', inplace=True)
RREAS_meta = df.copy()
df.head()
```

## Link Metadata files

- use unique Flyer2018 metadata sheet to add to Flyer 2018 otu table and get metadata for each "sequenced" sample

```python
#contains both Flyer2018 and RREAS samples
df = otu_all.copy()
df = df.T
df['Dataset'] = 'Merged2018'
df= df[['Dataset']]
df.index.rename('seqID', inplace=True)
#drop library ID and recreate sample name (unique Filter ID)
df['sample_name'] = df.index.str.split('_').str[:-1]
df['sample_name'] =df['sample_name'].str.join('_')

# Join with RREAS metadata first by seqID:
df = df.join(RREAS_meta)

# Now join with Flyer2018 metadata
df= df.reset_index().set_index('sample_name')

df = df.join(Flyer_meta, rsuffix= '_Flyer')
# Now have info for each Flyer sample
print(list(df))

#set seqID as index
df= df.reset_index().set_index('seqID')

print(directory + prefix + '_' + marker + '_meta_filtered.csv')
df.to_csv(directory + prefix + '_' + marker + '_meta_filtered.csv')
df.head()
```
