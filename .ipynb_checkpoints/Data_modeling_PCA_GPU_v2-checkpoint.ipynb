### Function to model the ASV count distributions

```python
def ranRelPct_df(df):
    def betaRow(row):
      ran_row = np.random.beta(row + 1, row.sum() - row + 1)
      return ran_row
    return df.apply(lambda x: betaRow(x), axis = 1, result_type='expand')
```

#### Run the function in a loop over both amplicon data sets and make a list of the two data frames

```python
asvs_list = []

for file_name in file_names:
    # Set the folder containing the data
    DATA_DIR = "./Data/"
    
    # Create the file path 
    file_path = DATA_DIR + file_name
    
    # Load the dataframe
    asvs = cudf.read_csv(file_name, index_col=0, sep='\t')
    
    # Model the ASV read counts
    asvs_modeled = ranRelPct_df(asvs)
    
    # Add the modeled df to the list
    asvs_list.append(asvs_modeled)
```

### Function for PCA

```python
def plot_pca_var(gdf):
    
    x_label = "component_number"
    y_label = "explained_variance"

    print(f"The input dataset has a shape of {gdf.shape}")
    
    # Identify the number or components
    n_components = gdf.shape[0]
    
    # Perform PCA
    pca = PCA(n_components = n_components)
    pca.fit_transform(gdf)
    
    # Calculate the proportion of the variance
    exp_var = pca.explained_variance_ratio_
    
    # Convert to percentage
    variance_array = exp_var.cumsum() * 100
    
    #Create a dataframe for plotting
    va_df = cudf.DataFrame({y_label:variance_array,x_label:range(1,len(variance_array)+1)})
    
    # Create title string
    title_str = f"Principal components of ASV count probabilities \n{file_name}"
    
    # Plot the data
    lp = sns.lineplot(data = va_df.to_pandas(), x = x_label, 
             y = y_label).set(title=title_str) 
    
    # Return the plot
    return lp
```

### Plot principal components of raw ASV count probabilities

```python
def scatter_plot_pca(file_name,n_components):
    
    # Set the folder containing the data
    DATA_DIR = "./Data/"
    
    file_path = DATA_DIR + file_name
    
    df = pd.read_csv(file_path, index_col=0, sep='\t')
    
    pca = PCA(n_components = n_components)
    
    pcs = pca.fit_transform(df) 

    pc1_values = pcs[:,0]
    pc2_values = pcs[:,1]
    
    # Create title string
    title_str = f"PCA, raw count probabilities\n{file_name}"
    
    sns.scatterplot(x=pc1_values, y=pc2_values).set(title=title_str)
    
```
