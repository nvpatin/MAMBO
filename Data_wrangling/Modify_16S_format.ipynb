# Modify 16S data format

```python
import pandas as pd
# import glob
```

### Set Locations

```python
directory = "./Data/merged_data/"
marker = '16S'
prefix =  'Merged2018'
```

# Import Data

```python
#asv
file = '/Users/kpitz/github/NOAA-NCAR-Hackathon/Data/merged_data/Merged2018_16S_asv_filtered.csv'
df = pd.read_csv(file)
df.set_index('ASV', inplace=True)
asv = df.copy()

#taxa
file = '/Users/kpitz/github/NOAA-NCAR-Hackathon/Data/merged_data/Merged2018_16S_taxa_filtered.csv'
df = pd.read_csv(file)
df.set_index('ASV', inplace=True)
taxa = df.copy()

#seq
file = '/Users/kpitz/github/NOAA-NCAR-Hackathon/Data/merged_data/Merged2018_16S_seq_filtered.csv'
df = pd.read_csv(file)
df.set_index('ASV', inplace=True)
seq = df.copy()
```

## Make numbered ASVs

```python
levels = list(taxa)
print(levels)

df = asv.copy()
df['tot'] = df.sum(axis=1)
df = df.sort_values('tot', ascending=False)
df = df.loc[df['tot']>0] #remove ASVs with 0 reads in dataset
df = df.reset_index()
df = df.rename(columns={'ASV':'Hash'})
df['ASV'] = df.index +1
df['ASV'] = 'ASV_' + df['ASV'].astype(str)
df = df.drop('tot', axis=1)
df.set_index('Hash', inplace=True)
#join with other dataframes, with new index
df = pd.concat([df, seq, taxa], axis=1)
df = df.reset_index()
#remove ASVs that were removed earlier
df = df.loc[df['ASV'].isna()==False]
df.set_index('ASV', inplace=True)
df = df.rename(columns={'index':'Hash'})
taxa_new = df[levels]
seq_new = df[['Hash', 'sequence']]
df.drop(levels,axis=1, inplace=True)
df.drop(['Hash', 'sequence'],axis=1, inplace=True)
asv_new = df.copy()
df
```

# Export using naming convention

```python
print(directory)
```

```python
#export to csv files for R plotting
dfs = [asv_new, taxa_new, seq_new]
names = ['otu', 'taxa', 'seq']
for df, name in zip(dfs,names):
    df.to_csv(directory + prefix + '_'+marker+'_'+name+'_filtered.csv')
    print(directory + prefix + '_'+marker+'_'+name+'_filtered.csv')
df.head()
```
