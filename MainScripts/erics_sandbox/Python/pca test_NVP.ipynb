```python
import pandas as pd
import numpy as np
import MarineDNA as md
import plotly.offline as pyo
import plotly.graph_objs as go
import plotly.express as px
```

```python
file1 = "../../../Data/Flyer2018_16S_table_counts.tsv"
asvs1 = pd.read_csv(file1, index_col=0, sep="\t")
```

```python
num_reps = 100

pca_list =  [md.doPCA(md.ranRelPct(asvs1)) for i in range(num_reps)]
```

### Loadings Summary

```python
loadings = [x["loadings"] for x in pca_list]

smry0 = md.loadingSummary(
    loadings, 
    pc = 0, 
    asv_labels = 
    asvs1.transpose().columns.values,
    z_thresh = 3,
    pct_outlier_thresh = 0.95
)
```

The positive loading ASVs on the first component

```python
smry0['positive']
```

The negative loading ASVs on the first component

```python
smry0['negative']
```

The first component is defined as a contrast between these two sets of ASVs

### Summarize loadings for all components

```python
smry_all = [
    md.loadingSummary(
        loadings, 
        pc = pc, 
        asv_labels = 
        asvs1.transpose().columns.values,
        z_thresh = 3,
        pct_outlier_thresh = 0.95
    ) for pc in range(len(loadings))
]
```

```python
smry_all[3]['positive']
```

```python
smry_all[3]['negative']
```

This extracts the defining ASVs for the third component from the smry_all list

```python
smry_all[0]['metrics']['loadings']
```
