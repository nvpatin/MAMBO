```python
test_counts = counts16s.iloc[:, 0:10]

p = md.ranRelPct(test_counts.to_numpy())
lo = np.log(p / (1 - p))
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca_fit = pca.fit(lo)
```

```python
loadings = np.transpose(pca_fit.components_)
print(loadings)
```

```python
def testFunc(mat):
    p = md.ranRelPct(mat)
    lo = np.log(p / (1 - p))
    pca = PCA()
    return np.transpose(pca.fit(lo).components_)

testFunc(test_counts.to_numpy())
```
