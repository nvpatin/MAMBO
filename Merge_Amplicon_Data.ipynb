# Merge Amplicon Datasets by ASV sequence

Join "Flyer2018" and "RREAS2018" datasets for 18S, COI, and 12S

```python
import pandas as pd
# import glob
```

### Set Locations

```python
directory = "./Data/"
markers = ['18S', 'COI', '12S']
# prefixes = ['RREAS2018', 'Flyer2018']
```

## Import Data

### RREAS2018

```python
marker = '18S'
prefix = 'RREAS2018'

# otu table
file = prefix +'_'+marker+'_otu_filtered.csv'
print(directory+file)
df = pd.read_csv(directory + file)
df = df.rename(columns= {'Unnamed: 0': 'ASV'})
df.set_index('ASV', inplace=True)
otu_all = df.copy()
print('Number ASVs:', len(df.index))

# taxa table
file = prefix +'_'+marker+'_taxa_filtered.csv'
df = pd.read_csv(directory+file)
df = df.rename(columns= {'Unnamed: 0': 'ASV'})
df.set_index('ASV', inplace=True)
taxa_all = df.copy()
print('Number ASVs:', len(df.index))

# sequence table
file = prefix +'_'+marker+'_seq_filtered.csv'
df = pd.read_csv(directory +file)
df = df.rename(columns= {'Unnamed: 0': 'ASV'})
df.set_index('ASV', inplace=True)
seq_all = df.copy()
print('Number ASVs:', len(df.index))

df.head()
```

```python
marker = '18S'
prefix = 'Flyer2018'

# otu table
file = prefix +'_'+marker+'_table_counts.tsv'
print(directory+file)
df = pd.read_csv(directory + file, sep='\t')
df.set_index('ASV ID', inplace=True)
otu_all = df.copy()
print('Number ASVs:', len(df.index))
df.head()
```
